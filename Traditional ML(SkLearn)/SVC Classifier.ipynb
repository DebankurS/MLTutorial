```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.datasets import make_blobs

x,y= make_blobs(n_samples=100, centers=3, n_features=3, cluster_std=0.8,
                  random_state=65)
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y)

plt.show()
```

```python
y
```

```python
from sklearn.svm import SVC

clf = SVC(kernel='linear')
clf.fit(x,y)
```

```python
w = clf.coef_[0]

a = -w[0] / w[1]

xx = np.linspace(min(x[:, 0]),max(x[:, 0]))

yy = a * xx - clf.intercept_[0] / w[1]

h0 = plt.plot(xx, yy, 'k-')

plt.scatter(x[:, 0], x[:, 1], c = y)

plt.show()
```
